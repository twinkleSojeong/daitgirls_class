```python
from datetime import datetime, timedelta

the_day = datetime(2021, 9, 13)
row_dict ={}

for days in [5, 10, 20, 60, 120, 240]:
    start = the_day - timedelta(days)
    end = the_day
    
    df = med[start:end] #특정기간
    acc_rets = df / df.iloc[0]-1.0
    row_dict[days] = acc_rets.iloc[-1]
```

```python
df_rets = pd.DataFrame(row_dict)
df_rets
```

```python
#전체 기간별 수익률
df_rets.mean()
```

```python
#효은님 코드
def color_negative_blue(val):
    color = 'blue' if val < 0 else 'black'
    return 'color: %s' % color

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: red' if v else '' for v in is_max]

df_rets.style.\
    applymap(color_negative_red).\
    apply(highlight_max)
```

```python
## 엑셀로 저장하기
df_rets.style.\
    applymap(color_negative_blue).\
    apply(highlight_max).\
    to_excel('건축_기간별수익률_2021_09_13.xlsx', engine='openpyxl')
```
